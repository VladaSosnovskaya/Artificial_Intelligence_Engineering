Импорт библиотеки NumPy с общепринятым сокращением

```python
import numpy as np
```

Примеры создания массивов: из обычных списков, с использованием функций zeros, ones, arange или аналогичных.

```python
arr_from_list = np.array([x for x in range(1, 21)])
print(f"Массив от 1 до 20 из обычного списка: {arr_from_list}")
arr_zeros = np.zeros(20)
print(f"Массив из 20 нулей: {arr_zeros}")
arr_ones = np.ones(20)
print(f"Массив из 20 единиц: {arr_ones}")
arr_arange = np.arange(1, 21, 2)
print(f"Массив от 1 до 20 с шагом 2: {arr_arange}")
```

Демонстрация атрибутов: вывод формы (shape) и типа данных (dtype) для одного из массивов.

```python
print(f"Форма: {arr_from_list.shape}, тип данных: {arr_from_list.dtype}")
```

Примеры индексирования и срезов: получение отдельного элемента, извлечение подмассива с помощью среза.

```python
print(f"Первый элемент массива: {arr_from_list[0]}")
print(f"Последний элемент массива: {arr_from_list[-1]}")
print(f"Срез от 1 до 4 элемента: {arr_from_list[1:4] }")
```

Примеры векторных операций и базовых статистик: поэлементное сложение или умножение массивов, вычисление средней величины (mean) или другой простой статистики.

```python
arr_1 = np.array([x for x in range(1, 11)]) # от 1 до 10
arr_2 = np.array([x for x in range(11, 21)]) # от 11 до 20
print(f"Поэлементраное сложение двух массивов {arr_1 + arr_2}") # 1+11, 2+12 и т.д
print(f"Поэлементраное умножение двух массивов {arr_1 * arr_2}") # 1*11, 2*12 и т.д
print(f"Средняя величина 1 массива {arr_1.mean()}") 
print(f"Средняя величина 2 массива {arr_2.mean()}") 
arr_3 = np.concatenate((arr_1, arr_2)) # от 1 до 20
print(f"Объединенный массив {arr_3}")
print(f"Средняя величина объединенного массива {arr_3.mean()}") 
```
