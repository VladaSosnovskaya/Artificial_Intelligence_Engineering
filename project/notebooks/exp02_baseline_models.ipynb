**Цель:**   
**Данные:** `data/train/`, `data/test/`   
**Подход:**   

```python
from src.constants import DATA_PATH
from src.data.loader import create_dataloaders
from src.models.baselines import prepare_flat_features, train_baseline_models
from sklearn.preprocessing import StandardScaler
from src.utils.experiment_logger import log_experiment

# Создаём DataLoader БЕЗ аугментаций для честного сравнения с классическими моделями
train_loader, val_loader, test_loader = create_dataloaders(DATA_PATH, batch_size=64)

X_train, y_train = prepare_flat_features(train_loader)
X_test, y_test = prepare_flat_features(test_loader)

# StandardScaler для логистической регрессии
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

baseline_results = train_baseline_models(X_train, y_train)

experiment_name = "baselines"
for model_name in baseline_results:
    model = baseline_results[model_name]
    preds = model.predict(X_test_scaled)

    training_cfg = {
        "experiment_name": experiment_name,
        "model_name": model_name,
        "scaler": "StandardScaler", 
        "augmentation": "None"
    }

    log_experiment(experiment_name, model_name, y_test, preds, training_cfg)
```
