**Цель:**   
**Данные:** `data/train/`, `data/test/`   
**Подход:**   

```python
from src.data.loader import create_dataloaders
from src.models.cnn import train_cnn, eval_cnn
from src.constants import DATA_PATH
from src.utils.experiment_logger import log_experiment

batch_size = 64
train_loader, val_loader, test_loader = create_dataloaders(DATA_PATH, batch_size=batch_size)

experiments = [
    {"model_name": "CNN_Baseline", "lr": 1e-3, "wd": 1e-4, "epochs": 50},
    {"model_name": "CNN_Improved", "lr": 5e-4, "wd": 5e-4, "epochs": 50}
]

experiment_name = "cnn_variants"
for exp in experiments:
    patience = 5
    model_name = exp["model_name"]
    model, hist = train_cnn(train_loader, val_loader, epochs=exp["epochs"], lr=exp["lr"], name=model_name, 
                            weight_decay=exp["wd"], patience=patience)
    y_true, y_pred = eval_cnn(model, test_loader)

    training_cfg = {
        "experiment_name": experiment_name,
        "model_name": model_name,
        "epochs": len(hist["train_loss"]),
        "lr": exp["lr"],
        "wd": exp["wd"],
        "patience": patience,
        "batch_size": batch_size,
        "augmentation": "None"
    }

    metrics = log_experiment(experiment_name, model_name, y_true, y_pred, training_cfg)
```
